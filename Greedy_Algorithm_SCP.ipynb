```python
R = set()
S = []
w = []
file1 = open('C:/Users/abiol/OneDrive/Documents/PHD Supervisory Meetings/SCP_Data.txt', 'r')
Lines = file1.readlines()
 
count = 0
# Strips the newline character
for line in Lines:
    count += 1
    f1 = line.strip() #this is a representation of your individual set
    f2 = f1.split(' ')
    #print(f2)
    uniqueline = set(f2)
    S.append(uniqueline)
    R.update(f2)
    w.append(1)
    #print(universe)
#print(Subs1)
```

```python
C = []
costs = []
```

```python
def findMin(Subsets, Universe):
    minCost = 99999.0
    minElement = -1
    for i, s in enumerate(Subsets):
        try:
            cost = weights[i]/(len(s.intersection(R)))
            if cost < minCost:
                minCost = cost
                minElement = i
        except:
            # Division by zero, ignore
            pass
    return Subsets[minElement], weights[minElement]

while len(Universe) != 0:
    S_i, cost = findMin(Subsets, Universe)
    C.append(S_i)
    R = R.difference(S_i)
    costs.append(cost)
```

```python
def findMin(S, R):
    minCost = 99999.0
    minElement = -1
    for i, s in enumerate(S):
        try:
            cost = w[i]/(len(s.intersection(R)))
            if cost < minCost:
                minCost = cost
                minElement = i
        except:
            # Division by zero, ignore
            pass
    return S[minElement], w[minElement]

while len(R) != 0:
    S_i, cost = findMin(S, R)
    C.append(S_i)
    R = R.difference(S_i)
    costs.append(cost)
```

```python
print("Cover: ", C)
print("Total Cost: ", sum(costs), costs)
```
